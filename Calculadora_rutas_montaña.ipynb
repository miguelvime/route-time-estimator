Para planificar mis rutas de montaña, quiero una calculadora que me permita ingresar la distancia total de la ruta en kilómetros y el desnivel acumulado en metros. La calculadora debe proporcionarme una estimación del tiempo total que me llevará completar la ruta. Para ello usará que el tiempo en llano será de 4 km/h, tardaremos en subir 300 metros de desnivel 1 hora, y en bajar 300 metros de desnivel 30 minutos. Añadirá además un 50% de tiempo extra para descansos y contingencias.

```python
distancia_horizontal = int(input("Inserta la distancia horizontal total de la ruta en metros"))
desnivel_positivo = int(input ("Inserta el desnivel positivo total de la ruta en metros"))
desnivel_negativo = int(input ("Inserta el desnivel negativo total de la ruta en metros"))
#Tengo distancias, ahora quiero los tiempos de cada una
tiempo_horizontal = distancia_horizontal * (1/4000) # horas horizontales en sistema decimal
tiempo_positivos = desnivel_positivo * (1/300) # en horas con decimales
tiempo_negativos = desnivel_negativo * (0.5/300) # En horas con decimales

tiempo_total_horas = tiempo_negativos + tiempo_horizontal + tiempo_positivos
tiempo_total_horas_descanso = 1.5 * tiempo_total_horas

print (f"Tardarás en hacer la ruta {tiempo_total_horas_descanso} horas")
```
